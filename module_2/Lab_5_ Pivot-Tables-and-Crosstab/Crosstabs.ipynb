### 2. Use crosstabs to show the number of movies per country

```python
# using groupby
movie.groupby("country")[["country"]].value_counts().to_frame().sort_values(by=0, ascending=False)
```

```python
# using crosstab
pd.crosstab(index=movie["country"], columns="count").sort_values(by="count", ascending=False)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(index=[movie["director_name"], movie["genres"], movie["country"]], columns="count")
```

### 4. Average duration of films grouped by color

```python
print("""

Unfortunately, crosstab is not the optimal choice for calculating the average
since it is mainly used for cross-tabulation of categorical variables.
However, you can still achieve the desired result using the groupby() or the pivot_table() function.

Here's how:
 
 """)
```

```python
movie.groupby("color")[["duration"]].mean()
```

```python
movie.pivot_table(index="color", values="duration", aggfunc="mean") #.reset_index(inplace=True)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
# using groupby

# get likes pser director (sum of actor's likes)
movie["totalLikes"] = movie["actor_1_facebook_likes"] + movie["actor_2_facebook_likes"] + movie["actor_3_facebook_likes"]

# average number of likes per director
movie.groupby("director_name")[["totalLikes"]].mean().sort_values(by="totalLikes", ascending=False)
```

```python
# using crosstab
pd.crosstab(index=movie["director_name"], columns="average_likes", values=movie["totalLikes"], aggfunc="mean").sort_values(by="average_likes", ascending=False)
```
