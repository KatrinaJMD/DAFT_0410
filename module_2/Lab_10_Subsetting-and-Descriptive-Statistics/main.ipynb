####  What is the average purchase price?

```python
# your code here
avg_purchasePrice = round(orders["amount_spent"].mean(), 2)

print(f"Average purchase price is €{avg_purchasePrice:,}")
```

#### What are the highest and lowest purchase prices? 

```python
# your code here
min_purchasePrice = round(orders["amount_spent"].min(), 2)
max_purchasePrice = round(orders["amount_spent"].max(), 2)

print(f"Lowest purchase price is €{min_purchasePrice}")
print(f"Highest purchase price is €{max_purchasePrice:,}")
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
totalOrders_perCustomer = orders.groupby("CustomerID", as_index=False)[["Quantity"]].sum()

ordersOver50 = totalOrders_perCustomer[totalOrders_perCustomer["Quantity"] > 50].sort_values(by="Quantity")
display(ordersOver50)

clients_ordersOver50 = ordersOver50["CustomerID"].unique()
print(f"No. of clients with total orders over 50 : {len(clients_ordersOver50):,}")
```
