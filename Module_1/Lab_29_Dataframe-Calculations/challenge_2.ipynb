````python
# your code here
# --------------

# making a copy of pokemon dataframe while only keeping the concerned columns, cause we never know..
variables = pokemon.iloc[:, 5:11].copy()
display(variables.head())
print(f"Making a copy of pokemon dataframe while only keeping the concerned columns, cause we never know..")

print("\n")

# creating a new column called HYPOTHESIS to test if our theory was right
variables["Hypothesis"] = variables.sum(axis=1)
display(variables.head())
print(f"Creating a new column called Hypothesis to test if our theory was right")

print("\n")

# go the extra mile because we're awesome, and create a new column called is_Match to confirm if the formula works for all pokemon
variables["is_Match"] = variables["Hypothesis"].isin(pokemon["Total"]).replace({True: "Yes", False: "No"})
display(variables.head())
print(f"Go the extra mile because we're awesome, and create a new column called is_Match to confirm if the formula works for all pokemon")

print("\n")

# count the number of matches
print(f"\nNumber of rows in our pokemon data : {len(pokemon)}")
print(f"Number of rows in our variables data : {len(variables)} (it shouldn't change from the pokemon data)")
print(f"Therefore, we should have {len(pokemon)} matches if our theory was right!\n")
display(variables["is_Match"].value_counts().to_frame())

print("\n")

# automate results
if len(variables.loc[variables["is_Match"] == "Yes"]) == len(pokemon):
    print("RESULT : Yeah, Total is just a sum of all those columns")
else:
    print("RESULT : Ha, dang it..")
````

## Problem Solving Iteration 2

Now that we have consolidated the abilities fields, we can update the problem statement. The new problem statement is:

### Which pokemon type is most likely to have the highest `Total` value?

In the updated problem statement, we assume there is a certain relationship between the `Total` and the pokemon type. But we have two *type* fields (`Type 1` and `Type 2`) that have string values. In data analysis, string fields have to be transformed to numerical format in order to be analyzed. 

In addition, keep in mind that `Type 1` always has a value but `Type 2` is sometimes empty (having the `NaN` value). Also, the pokemon type we choose may be either in `Type 1` or `Type 2`.

Now our expectation is:

#### `Type 1` and `Type 2` string variables need to be converted to numerical variables in order to identify the relationship between `Total` and the pokemon type.

The information we need to collect is:

#### How to convert two string variables to numerical?

Let's address the first question first. You can use a method called **One Hot Encoding** which is frequently used in machine learning to encode categorical string variables to numerical. The idea is to gather all the possible string values in a categorical field and create a numerical field for each unique string value. Each of those numerical fields uses `1` and `0` to indicate whether the data record has the corresponding categorical value. A detailed explanation of One Hot Encoding can be found in [this article](https://hackernoon.com/what-is-one-hot-encoding-why-and-when-do-you-have-to-use-it-e3c6186d008f). You will formally learn it in Module 3.

For instance, if a pokemon has `Type 1` as `Poison` and `Type 2` as `Fire`, then its `Poison` and `Fire` fields are `1` whereas all other fields are `0`. If a pokemon has `Type 1` as `Water` and `Type 2` as `NaN`, then its `Water` field is `1` whereas all other fields are `0`.

#### In the next cell, use One Hot Encoding to encode `Type 1` and `Type 2`. Use the pokemon type values as the names of the numerical fields you create.

The new numerical variables you create should look like below:

![One Hot Encoding](../images/one-hot-encoding.png)

````python
# your code here
# --------------

# encode Type 1 and visualize
type1_encoded = pd.get_dummies(pokemon['Type 1'])
display(type1_encoded.head())
print("TYPE 1")

# encode Type 2 and visualize
type2_encoded = pd.get_dummies(pokemon['Type 2'])
display(type2_encoded.head())
print("TYPE 2")

# fuse both types in a single dataframe and treat NaN values as 0
combined_types = type1_encoded.add(type2_encoded, fill_value=0)
display(combined_types.head())
print("COMBINED TYPES\n")

print("-" * 111)
print("""
NOTE TO READERS :
\n
In pandas, the axis parameter is used to specify the direction of an operation across the DataFrame.
The two possible values for axis are 0 and 1:

axis=0: This represents the vertical axis,
        meaning the operation will be performed along the index (rows).

When you use axis=0, you are telling pandas to apply the operation to each column individually, along the rows.
In other words, the operation will be performed column-wise.
This is the default value for most pandas operations.

axis=1: This represents the horizontal axis,
        meaning the operation will be performed along the columns (column-wise).
        
When you use axis=1, you are telling pandas to apply the operation to each row individually, along the columns.
In other words, the operation will be performed row-wise.

KATRINA J, Author.
""")
print("-" * 111 + "\n")

# join original data with the encoded data
print("POKEMON DATA CONCATENATED WITH ENCODED DATA")
pokemon_encoded = pd.concat([pokemon[["Type 1", "Type 2"]], combined_types], axis=1)
display(pokemon_encoded)
````

**NOTE TO SELF** ***(for future reference)***

In pandas, both **`.corr()`** and **`.corrwith()`** methods are used to compute correlation, but they **serve different purposes**:

> **`.corr()`** <br>
This method is called on a DataFrame and **calculates the pairwise correlation between columns or rows** of the DataFrame. By default, it computes the correlation between columns **(axis=0)**.

If you want to compute the correlation between rows, you can set the axis parameter to 1. The method returns a correlation matrix with the correlation coefficients between each pair of columns or rows.

> **EXAMPLE :**

```python
import pandas as pd

data = {'A': [1, 2, 3],
        'B': [2, 3, 4],
        'C': [3, 4, 5]}

df = pd.DataFrame(data)

correlation_matrix = df.corr()
print(correlation_matrix)
```

> **`corrwith()`**<br>
This method is called on a DataFrame and **calculates the correlation between its columns or rows** ***AND*** **another DataFrame**, **Series**, or **array**.

It is useful when you want to compute the correlation between the columns or rows of two different DataFrames or between the columns of a DataFrame and a Series.

> **EXAMPLE :**

```python
import pandas as pd

data1 = {'A': [1, 2, 3],
         'B': [2, 3, 4]}
data2 = {'C': [3, 4, 5],
         'D': [4, 5, 6]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# correlation between columns of df1 and columns of df2
correlation = df1.corrwith(df2)
print(correlation)

```

````python
# your code here
print("NOOB VERSION WITH LOTS OF LINES (BUT AT LEAST SOMEHOW STRUCTURED)")

print("\n")
print("-" * 135)

# reuse combined_types data to isolate ENCODED fields, and join the TOTAL column from the pokemon data
combined_types["Total"] = pokemon["Total"]
display(combined_types.head())
print("JOIN TOTAL COLUMN TO THE ENCODED TYPES")

print("\n")
print("-" * 135)

# calculate correlation between TOTAL and ENCODED fields
# column TOTAL is dropped so we won't face issues when identifying the #1 Pokémon type
print("RESULTS OF (UNSORTED) CORRELATIONS (TOTAL COLUMN IS DROPPED FOR RANKING PURPOSES)")
correlations = combined_types.corr()["Total"].drop("Total")
display(correlations)

print("\n")
print("-" * 135)

# ranking correlations from highest to lowest
print("CORRELATIONS RANKED (HIGHEST TO LOWEST)")
ranked_correlations = correlations.sort_values(ascending=False).to_frame()
ranked_correlations = ranked_correlations
display(ranked_correlations)

print("\n")
print("-" * 135)

# identify #1 Pokémon type most likely to have the highest TOTAL
theBEAST = ranked_correlations["Total"].idxmax()
display(ranked_correlations.loc[ranked_correlations["Total"] == ranked_correlations["Total"].max()])
print(f"#1 POKEMON TYPE MOST LIKELY TO HAVE THE HIGHEST TOTAL IS {theBEAST}")
````

# Bonus Question

Say now you can choose both `Type 1` and `Type 2` of the pokemon. In order to receive the best pokemon, which types will you choose?

````python
print("""
To determine the best combination of Type 1 and Type 2 for a Pokémon,
we can analyze the correlations between the encoded type fields and the "Total" column,
which we calculated earlier.

However, this approach considers only the individual correlation of each type with the "Total" column,
and does not take into account potential interactions between the two types.

A more accurate approach would be to examine the average "Total" values for each unique combination of Type 1 and Type 2 in the dataset.
""")
````

````python
# your code here
print("NOOB VERSION WITH LOTS OF LINES (BUT AT LEAST SOMEHOW STRUCTURED) :\n")

TypesGroupedBy = pokemon.groupby(["Type 1", "Type 2"]).mean(numeric_only=True)["Total"]
BestCombo = TypesGroupedBy.idxmax()
MaxTotal = TypesGroupedBy.max()

print(f"The best combination of Type 1 and Type 2 is {BestCombo} with an average Total of {MaxTotal}")
````
