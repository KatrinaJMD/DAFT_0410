## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# --------------

# total amount_spent per customer
print("SUB PROBLEM 1 : SUM SPENT PER CUSTOMER")
TotalSpent = orders.groupby("CustomerID")["amount_spent"].sum()
display(TotalSpent.to_frame())
```

```python
# customers above 95th percentile and between 75th and 95th percentile
print("SUB PROBLEM 2 : IDENTIFY CUSTOMER STATUS\n")

vip = TotalSpent.quantile(0.95) # 95% of the data is below the result
preffered = TotalSpent.quantile(0.75) # 75% of the data is below the result

print(f"95th PERCENTILE CORRESPONDS TO A TOTAL AMOUNT SPENT OF {float(vip)}")
print(f"BETWEEN 75th AND 95th PERCENTILE CORRESPONDS TO A TOTAL AMOUNT SPENT OF {float(preffered)}")

# filter customers based on quantile ranges
vipCustomers = TotalSpent[TotalSpent >= float(vip)]
preferredCustomers = TotalSpent[(TotalSpent >= float(preffered)) & (TotalSpent < float(vip))]

display(vipCustomers.to_frame())
print(f"VIP CUSTOMERS")

print("\n")

display(preferredCustomers.to_frame())
print(f"PREFERRED CUSTOMERS")
```

```python
# customers above 95th percentile and between 75th and 95th percentile
print("SUB PROBLEM 3 : LABEL CUSTOMERS BASED ON TOTAL AMOUNT SPENT\n")

# function that sets customer staus based on the quantile thresholds
def customerLabel(spend):
    if spend >= float(vip):
        return "VIP"
    elif spend >= float(preffered):
        return "Preferred"
    else:
        return "Regular"

# apply function to TotalSpent Series or DataFrame
CustomerStatus = TotalSpent.apply(customerLabel)
display(CustomerStatus.to_frame())
print(f"CUSTOMER STATUS PER CUSTOMER ID\n")

# create new DataFrame with CustomerID and CustomerStatus columns
CustomerStatus_df = pd.DataFrame(CustomerStatus).reset_index().rename(columns={"amount_spent": "CustomerStatus"})

# merge CustomerStatus_df with the original"'order"' DataFrame
OrderInfo = orders.merge(CustomerStatus_df, on="CustomerID")

# show result
display(OrderInfo.head())
print(f"CustomerStatus COLUMN ADDED TO THE ORIGINAL ORDERS DATAFRAME WITH THE MATCHING STATUS")
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
Method1 = OrderInfo.loc[OrderInfo["CustomerStatus"] == "VIP"][["CustomerStatus", "Country"]].value_counts().sort_values(ascending=False).to_frame()
Method2 = OrderInfo.loc[OrderInfo["CustomerStatus"] == "VIP"][["Country", "CustomerStatus"]].value_counts().sort_values(ascending=False).to_frame()

print(f"METHOD 1")
display(Method1)

print(f"\nMETHOD 2")
display(Method2)

# if we only want to show THE country that has tHE MOST VIP customers
print(f"\nIf we only want to show THE country that has THE MOST VIP customers\n")

print(f"METHOD 1")
display(Method1.head(1))

print(f"\nMETHOD 2")
display(Method2.head(1))
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# --------------

# isolate concerned statuses so we're tranquil xD
CustomerCombo = OrderInfo.loc[(OrderInfo["CustomerStatus"] == "VIP") | (OrderInfo["CustomerStatus"] == "Preferred")]

# check if eveything went well as planned so we won't look stupid moving forward
print(f"CUSTOMER STATUSES/LABELS PRESENT IN OUR CustomerCombo DATA : {CustomerCombo.CustomerStatus.unique()}")
print("-" * 96)

# count VIP and Preffered customers combined epr country
TopCountry = CustomerCombo.groupby('Country')['CustomerID'].nunique().sort_values(ascending=False)
display(TopCountry.to_frame().head(1))

# dedicated result
print(f"\nCOUNTRY WITH THE MOST VIP AND PREFFERED CUSTOMERS COMBINED IS {TopCountry.index[0]} WITH A TOTAL OF {TopCountry.iloc[0]}")
print("-" * 96)

# full overview of all countries
print(f"\nFULL VIEW (IF YOU'RE CURIOUS, WHICH YOU ALWAYS SHOULD :P)")
display(TopCountry.to_frame())
```
