#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
# check the concerned column
pokemon["Name"].loc[pokemon["Name"].str.contains("mega", case=False)].head().to_frame()
```

```python
# your code here
print(f"METHOD 1 : applying Lambda")
pokemon["Name_M1"] = pokemon["Name"].apply(lambda name: "Mega " + name.split("Mega")[-1].strip() if "Mega" in name else name)

print(f"VERIFY RESULTS : see last column named name_M1 (reference to METHOD 1)")
print("-" * 132 + "\n")
pokemon.loc[pokemon["Name"].str.contains("mega", case=False)].head()
```

```python
print(f"METHOD 2 : using function")

def clean_mega_names(name):
    if "Mega" in name:
        to_fix = name.split("Mega")
        fixed = "Mega " + to_fix[-1].strip()
        return fixed
    else:
        return name

pokemon["Name_M2"] = pokemon["Name"].apply(clean_mega_names)

print(f"VERIFY RESULTS : see last column named name_M2 (reference to METHOD 2)")
print("-" * 132 + "\n")
pokemon.loc[pokemon["Name"].str.contains("mega", case=False)].head()
```

#### Create a new column called `A/D Ratio` whose value equals to `Attack` devided by `Defense`.

For instance, if a pokemon has the Attack score 49 and Defense score 49, the corresponding `A/D Ratio` is 49/49=1.

```python
# check the concerned column
pokemon[["Attack", "Defense"]].head()
```

```python
# your code here
pokemon["A/D.Ratio"] = round(pokemon["Attack"] / pokemon["Defense"], 2)
pokemon.head()
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
# your code here

# simplified version
# pokemon["Combo Type"] = pokemon["Type 1"] + "-" + pokemon["Type 2"]

# one-liner, cause we're awesome!
pokemon["Combo Type"] = pokemon.apply(lambda pokemonType: pokemonType["Type 1"] + "-" + pokemonType["Type 2"] if pd.notna(pokemonType["Type 2"]) else pokemonType["Type 1"], axis=1)
pokemon.head()
```

#### Identify the pokemon whose `A/D Ratio` are among the top 5.

```python
Top5 = pokemon.sort_values(by = "A/D.Ratio", ascending=False).head()
Top5
```

#### For the 5 pokemon printed above, aggregate `Combo Type` and use a list to store the unique values.

Your end product is a list containing the distinct `Combo Type` values of the 5 pokemon with the highest `A/D Ratio`.

```python
# your code here
Top5_combo = list(Top5["Combo Type"].unique())
Top5_combo
```

#### For each of the `Combo Type` values obtained from the previous question, calculate the mean scores of all numeric fields across all pokemon.

Your output should look like below:

![Aggregate](../images/aggregated-mean.png)

```python
# your code here

# check if the Top5 combo list is here
checkList = pokemon.loc[pokemon["Combo Type"].isin(Top5_combo)]["Combo Type"].value_counts().to_frame()
display(checkList)

AvgScore = pokemon.loc[pokemon["Combo Type"].isin(Top5_combo)].groupby("Combo Type").mean(numeric_only=True).rename(columns={0:"AvgScores"})
np.round(AvgScore, decimals=2)
```
